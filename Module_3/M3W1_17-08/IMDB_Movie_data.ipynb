```python
def rating_category(rating):
    if rating >= 7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'

data['Rating Category'] = data['Rating'].apply(rating_category)
data[['Title', 'Director', 'Rating', 'Rating Category']].head()
```
