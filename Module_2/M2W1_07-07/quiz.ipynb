```python
# Câu hỏi 10
import numpy as np

def maxx(x, y ) :
    if x >= y :
        return x
    else :
        return y
a = np.array ([5, 7, 9, 8, 6, 4, 5])
b = np.array ([6, 3, 4, 8, 9, 7, 1])

pair_max = np.vectorize(maxx, otypes =[float])
print (pair_max(a, b))
```

# Bài tập 3: Phân tích dữ liệu dạng bảng

```python
# Download data
# ! gdown 1iA0WmVfW88HyJvTBSQDI5vesf-pgKabq

import pandas as pd
df = pd.read_csv('./advertising.csv')

data = df.to_numpy()
```

```python
df
```

```python
# Câu hỏi 15
import numpy as np

sales_lst = data[:, -1]

sales_max = np.max(sales_lst)
index = np.where(sales_lst == sales_max)

print('Max: ' , sales_max , 'Index: ', index)
```

```python
# Câu hỏi 16
import numpy as np

TV_lst = data[:, 0]
np.average(TV_lst)
```

```python
# Câu hỏi 17

import numpy as np

sales_lst = data[:, -1]

np.sum(sales_lst >= 20)
```

```python
# Câu hỏi 18

import numpy as np

data_new = data[np.where(data[:, -1] >= 15)]
np.average(data_new[:, 1])
```

```python
# Câu hỏi 19

import numpy as np

average_Newspaper = np.average(data[:, 2])

data_new = data[np.where(data[:, 2] > average_Newspaper)]

data_new[:, -1].sum()
```

```python
# Câu hỏi 20

import numpy as np

A = np.average(data[:, -1])

scores = np.where(data[:, -1] > A, 'Good', np.where(data[:, -1] < A, 'Bad', 'Average'))

scores[7:10]
```

```python
# Câu hỏi 21

import numpy as np

A = np.average(data[:, -1])

distances = np.abs(data[:, -1] - A)

idx_min_distace = np.argmin(distances)

A = data[:, -1][idx_min_distace]

scores = np.where(data[:, -1] > A, 'Good', np.where(data[:, -1] < A, 'Bad', 'Average'))

scores[7:10]
```
