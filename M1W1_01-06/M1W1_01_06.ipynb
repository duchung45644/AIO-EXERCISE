```python
def exer1(tp, fp, fn):
  if(type(tp) != int):
    print('tp must be int')
    return
  if(type(fp) != int):
    print('fp must be int')
    return
  if(type(fn) != int):
    print('fn must be int')
    return

  if(tp <= 0 or fp <= 0 or fn <= 0):
    print('tp and fp and fn must be greater than zero')
    return
  precision = tp / (tp + fp)
  recall = tp / (tp + fn)
  f1_score = 2*precision*recall / (precision+recall)

  print('precision: ' , precision)
  print('recall: ' , recall)
  print('f1_score: ' , f1_score)
```

```python
exer1(2,3,4)
```

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
import math

def elu(x, alpha=1):
    return x if x >= 0 else alpha * (math.exp(x) - 1)
```

```python
def is_number(n) :
  try :
    float ( n )
  except ValueError :
    return False
  return True
```

```python
def exer2():
  x = input('Input x: ')
  if(not is_number(x)):
    print('x must be a number')
    return
  else:
    x = float(x)

  func_name = input('Input activation Function ( sigmoid | relu | elu ): ')

  if(func_name == 'sigmoid'):
    result = sigmoid(x)
  elif(func_name == 'relu'):
    result = relu(x)
  elif(func_name == 'elu'):
    result = elu(x, 0.01)
  else:
    print(func_name + ' is not supported')
    return


  print(f'{func_name}: f({x}) = {result}')
```

```python
import math

def MAE_s(predict, target):
  return math.fabs(predict - target)
def MSE_s(predict, target):
  return (predict - target)**2
```

```python
import math
import random

def MAE_s(predict, target):
    return math.fabs(predict - target)

def MSE_s(predict, target):
    return (predict - target)**2

def exer3():
  num_samples = input('Input number of samples (integer number) which are generated: ')
  if not num_samples.isnumeric():
    print('Number of samples must be an integer number')
    return

  num_samples = int(num_samples)

  loss_name = input('Input loss name (MSE | MAE | RMSE): ')
  if loss_name == 'MSE':
    loss_func = MSE_s
  elif loss_name == 'MAE':
    loss_func = MAE_s
  elif loss_name == 'RMSE':
    loss_func = MSE_s

  result = 0
  for i in range(num_samples):
    predict = random.uniform(0, 10)
    target = random.uniform(0, 10)

    loss = loss_func(predict, target)

    print(f'sample-{i}')
    print(f'\tpred: {predict}, target: {target}, loss: {loss}')

    result += loss

  result = result/num_samples
  if loss_name == 'RMSE':
    result =  math.sqrt(result)
  print(f'Final {loss_name}', result)

exer3()
```

```python
def factorial(n):
  if n==0:
    return 1
  else:
    return n * factorial(n-1)

def sin(x, n):
  result = 0
  for i in range(n):
    result += (-1)**i * (x**(2*i + 1) / factorial(2*i + 1) )
  return result

def cos(x, n):
  result = 0
  for i in range(n):
    result += (-1)**i * (x**(2*i) / factorial(2*i) )
  return result

def sinh(x, n):
  result = 0
  for i in range(n):
    result += (x**(2*i + 1) / factorial(2*i + 1) )
  return result

def cosh(x, n):
  result = 0
  for i in range(n):
    result += (x**(2*i) / factorial(2*i) )
  return result

print(sin(3.14, 10))
print(cos(3.14, 10))
print(sinh(3.14, 10))
print(cosh(3.14, 10))
```

```python
def MD_nRE_s(y, y_hat, n, p):
  return (y ** (1/ n ) - y_hat ** (1/ n )) ** p

def MD_nRE(y, y_hat, m, n, p):
  result = 0
  for i in range(m):
    result += MD_nRE_s(y, y_hat, n, p)

  return result
```

```python
MD_nRE(100, 99.5, 1, 2, 1)
MD_nRE(50, 49.5, 1, 2, 1)
```
